```python
#sklearn からdataset を利用する
from sklearn import datasets
#pandas パッケージを利用する
import pandas as pd
#load_iris でデータセットのサンプルを読み込む
dataset = datasets.load_iris()
#DataFrame に代入する
dataset2=pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns=dataset.feature_names
#irisの種類を取り出す　0:setosa 1:versicolor 2:virginica
dataset3=pd.DataFrame(dataset.target)
#列名を任意に指定する
dataset3.columns=['target']
#ふたつのデータセットを縦につなぎ合わせる
dataset4=pd.concat([dataset3, dataset2], axis=1)
#結果を出力する
dataset4
```

```python
#sklearn からdataset を利用する
from sklearn import datasets
#pandas パッケージを利用する
import pandas as pd
#load_wine でデータセットのサンプルを読み込む
dataset = datasets.load_wine()
#DataFrame に代入する
dataset2=pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns=dataset.feature_names
#wineのクラスを取り出す　0:setosa 1:versicolor 2:virginica
dataset3=pd.DataFrame(dataset.target)
#列名を任意に指定する
dataset3.columns=['Class']
#ふたつのデータセットを縦につなぎ合わせる
dataset4=pd.concat([dataset3, dataset2], axis=1)
#結果を出力する
dataset4
```
